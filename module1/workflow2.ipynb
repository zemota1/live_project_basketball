```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

```python
df = pd.read_csv("data_convertion.csv")
df.head()
```

```python
mask = np.random.rand(len(df)) < 0.8
train_df = df[mask]
test_df = df[~mask]
```

```python
train_df.shape, test_df.shape
```

```python
xx_train, yy_train = train_df['Celsius'], train_df['Fahrenheit']
xx_test, yy_test = test_df['Celsius'], test_df['Fahrenheit']
```

```python
model = keras.Sequential([
    layers.Dense(20, activation='relu', input_shape=[1]),
    layers.Dense(1)
  ])

optimizer = tf.keras.optimizers.Adam()

model.compile(loss='mse',
            optimizer=optimizer,
            metrics=['mae', 'mse'])
```

```python
model.summary()
```

```python
model.fit(
  xx_train, yy_train,
  epochs=100, validation_split = 0.2, verbose=1)
```

```python
yy_pred = model.predict(xx_test)
```

## Prediction

```python
for i in range(len(xx_test)):
    print(f"Celsius {list(xx_test)[i]}; Predicted {list(yy_pred)[i]}; True {list(yy_test)[i]}")
```
