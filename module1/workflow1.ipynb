# Manipulating Data with pandas

```python
import numpy as np
import pandas as pd

from datetime import datetime
```

```python
zeros = np.zeros((36, 4))
```

```python
df = pd.DataFrame(data=zeros, columns=['first_day_of_each_month', 'random_numbers1', 'random_numbers2', "difference"])
```

```python
df.index = pd.date_range('2018-01-01','2021-01-01', 
              freq='1M').tolist()
```

```python
df["first_day_of_each_month"] = [each_date.month for each_date in list(df.index)]
```

```python
df["random_numbers1"] = np.random.randint(1, 13, df.shape[0])
df["random_numbers2"] = np.random.randint(1, 13, df.shape[0])
```

```python
df['difference'] = df["random_numbers2"] - df["random_numbers1"]
```

```python
df_2018 = df.loc[df.index <= '2018-12-31']
df_2019 = df.loc[(df.index >= '2019-01-31') & (df.index <= '2019-12-31')]
df_2020 = df.loc[df.index >= '2020-01-01']
```
